```python
data["year"]=pd.to_datetime(data["year"])
```

```python
# Distribution of variables
```

```python

sns.displot(data["mass"],kind="ecdf")
sns.set_theme("talk")
sns.set_style("ticks")
```

```python
sns.displot(data["mass"],kind="kde",fill=True,color="black")
#sns.set("poster")
sns.set_style("white")
```

```python
sns.displot(data["mass"],kind="hist",fill=True,bins=15,color="r")
sns.set("poster")
sns.set_style("whitegrid")
```

# relationship between numerical variable

```python
sns.relplot(x=data["mass"],y=data["distance"],hue=data["method"],kind="scatter")
```

```python
sns.relplot(x=data["mass"],y=data["year"],hue=data["method"],kind="line",color="red")
sns.set("paper")
```

# relationship between categorical and numerical variable


```python
sns.catplot(x=data["method"],y=data["mass"], hue=data["number"],kind="bar")
```

```python
sns.catplot(x=data["method"],y=data["distance"], hue=data["number"],kind="box")
```

```python
sns.catplot(x=data["method"],y=data["distance"], hue=data["number"],kind="violin")
```

```python
sns.catplot(x=data["method"],y=data["distance"], hue=data["number"],kind="point")
```

```python
sns.catplot(x=data["method"],y=data["mass"], hue=data["number"],kind="strip")
```

```python
sns.catplot(x=data["method"],y=data["mass"], hue=data["number"],kind="swarm")
```
